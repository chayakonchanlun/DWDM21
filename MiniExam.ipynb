## **ข้อ 0 หารัฐที่มีจำนวนเครื่องบินขาเข้ามากที่สุดและน้อยที่สุด**

````python
merged_table = airports.merge(flights,how='left',left_on='รหัสสนามบิน',right_on='รหัสสนามบินต้นทาง')
merged_table #(เชื่อมตารางโดยใช้ index ของ รหัสสนามบินเเละรหัสสนามบินต้นทาง)
````

ตรวจสอบ missing


## ทำการลบ missing โดยใช้ **Dropna**

````python
merged_table.isnull().any()
````

````python
merged_table_dropna = merged_table.dropna()
merged_table_dropna
````

````python
merged_table_dropna.shape #จะเห็นได้ว่าหลังจากdropna มีข้อมูลของเเถวหายไปเป็นเเถวที่มีช่องว่าง Nan
````

ขาเข้ามากสุด เเละ ขาเข้าน้อยสุด

````python
total = merged_table_dropna.groupby('รัฐ')['ความล่าช้าขาเข้า'].max()
total = pd.DataFrame(total)
total.reset_index()  #นำตารางที่เป็น dropna มาเลือกเเค่คอลัมที่เราต้องการ
````

````python
max(total['ความล่าช้าขาเข้า'])
````

````python
total.describe()
````

````python
total[total['ความล่าช้าขาเข้า'] == 1845]  # ดั้งนั้นรัฐที่มีจำนวนเที่ยวบินขาเข้ามากที่สุดคือ รัฐ NV
````

````python
total_min = merged_table_dropna.groupby('รัฐ')['ความล่าช้าขาเข้า'].min()
total_min = pd.DataFrame(total_min)
total_min.reset_index()  #นำตารางที่เป็น dropna มาเลือกเเค่คอลัมที่เราต้องการ
````

````python
total_min.describe()
````

````python
min(total_min['ความล่าช้าขาเข้า'])
````

````python
total_min[total_min['ความล่าช้าขาเข้า'] == -94.0]
````

ตอบจำนวนเครื่องบิน ขาเข้ามากที่สุดคือรัฐ NV 1845 และรัฐที่มีเที่ยวบินขาเข้าน้อยที่สุดคือรัฐ RI -94

## **ข้อ 3 สายการบินไหนมีเที่ยวบินมากที่สุดและน้อยที่สุด**

````python
merged_table = flights.merge(airports,how='left',left_on='รหัสสนามบินต้นทาง',right_on='รหัสสนามบิน')
merged_table
````

## จากตาราง merged_table จะหาสายการบินมากที่สุด 


````python
merged_table.groupby('สายการบิน').count()
````

ให้ค่า sort values เป็น วันในเดือน โดยให้ค่า ascending เป็น False เพื่อให้เรียงจากมากลงไปน้อย

````python
summary = merged_table.groupby('สายการบิน').count().sort_values(by=['วันในเดือน'],ascending=False)
summary
````

จากผลลัพธ์ที่ได้นำมาหาค่า Max และ Min เพื่อหาเที่ยวบินที่มากสุดแะเที่ยวบินน้อยสุดจะได้

````python
max = summary.iloc[:1,:1]
max
````

พบว่าสายการบิน WN มีเที่ยวบินที่มากที่สุด =575739 ครั้ง


````python
min = summary.iloc[15:16,:1]
min
````

พบว่าสายการบิน WN มีเที่ยวบินที่น้อยที่สุด =17432 ครั้ง
